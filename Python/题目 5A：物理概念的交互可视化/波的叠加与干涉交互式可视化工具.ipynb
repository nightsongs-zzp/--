```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, VBox, Label
import warnings
warnings.filterwarnings('ignore')

# è®¾ç½®ä¸­æ–‡å­—ä½“ï¼ˆå¦‚æœç³»ç»Ÿæ”¯æŒï¼‰
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

def wave_superposition_interference(A1=1.0, A2=1.0, phi=0.0, wavelength=1.0):
    """
    å±•ç¤ºä¸¤ä¸ªæ³¢çš„å åŠ ä¸å¹²æ¶‰
    
    å‚æ•°ï¼š
    A1: æ³¢1çš„æŒ¯å¹…
    A2: æ³¢2çš„æŒ¯å¹…
    phi: æ³¢2ç›¸å¯¹äºæ³¢1çš„ç›¸ä½å·®ï¼ˆå¼§åº¦ï¼‰
    wavelength: æ³¢é•¿
    """
    
    # åˆ›å»ºç©ºé—´åæ ‡
    x = np.linspace(0, 3 * wavelength, 1000)
    k = 2 * np.pi / wavelength  # æ³¢æ•°
    
    # è®¡ç®—ä¸¤ä¸ªæ³¢
    wave1 = A1 * np.sin(k * x)
    wave2 = A2 * np.sin(k * x + phi)
    wave_sum = wave1 + wave2
    
    # åˆ›å»ºå›¾å½¢
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
    
    # ç»˜åˆ¶ç¬¬ä¸€ä¸ªæ³¢
    ax1.plot(x, wave1, 'b-', linewidth=2, label=f'æ³¢1: A={A1:.2f}')
    ax1.plot(x, wave2, 'r-', linewidth=2, alpha=0.7, label=f'æ³¢2: A={A2:.2f}, Ï†={phi:.2f} rad')
    ax1.plot(x, wave_sum, 'g-', linewidth=3, label=f'å åŠ æ³¢: æŒ¯å¹…â‰ˆ{np.max(np.abs(wave_sum)):.2f}')
    
    ax1.set_xlabel('ä½ç½® (x)')
    ax1.set_ylabel('æŒ¯å¹…')
    ax1.set_title('æ³¢çš„å åŠ ')
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='upper right')
    ax1.set_ylim(-(A1 + A2 + 0.5), A1 + A2 + 0.5)
    
    # ç»˜åˆ¶å¹²æ¶‰å¼ºåº¦å›¾
    interference_intensity = wave_sum**2  # å¼ºåº¦ä¸æŒ¯å¹…å¹³æ–¹æˆæ­£æ¯”
    ax2.plot(x, interference_intensity, 'purple', linewidth=2)
    ax2.fill_between(x, 0, interference_intensity, alpha=0.3, color='purple')
    
    ax2.set_xlabel('ä½ç½® (x)')
    ax2.set_ylabel('å¼ºåº¦ (I âˆ AÂ²)')
    ax2.set_title('å¹²æ¶‰å¼ºåº¦åˆ†å¸ƒ')
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim(0, (A1 + A2)**2 + 0.5)
    
    # æ ‡è®°ç›¸é•¿å’Œç›¸æ¶ˆå¹²æ¶‰åŒºåŸŸ
    # å¯»æ‰¾æå¤§å€¼å’Œæå°å€¼
    max_indices = np.where(np.diff(np.sign(np.diff(interference_intensity))) < 0)[0] + 1
    min_indices = np.where(np.diff(np.sign(np.diff(interference_intensity))) > 0)[0] + 1
    
    if len(max_indices) > 0:
        for idx in max_indices[:3]:  # åªæ ‡è®°å‰3ä¸ªæå¤§å€¼
            ax2.plot(x[idx], interference_intensity[idx], 'ro', markersize=8)
            ax2.annotate('ç›¸é•¿å¹²æ¶‰', xy=(x[idx], interference_intensity[idx]),
                        xytext=(10, 10), textcoords='offset points',
                        color='red', fontweight='bold')
    
    if len(min_indices) > 0:
        for idx in min_indices[:3]:  # åªæ ‡è®°å‰3ä¸ªæå°å€¼
            ax2.plot(x[idx], interference_intensity[idx], 'bo', markersize=8)
            ax2.annotate('ç›¸æ¶ˆå¹²æ¶‰', xy=(x[idx], interference_intensity[idx]),
                        xytext=(10, -15), textcoords='offset points',
                        color='blue', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    # ç‰©ç†åŸç†è¯´æ˜
    print("="*60)
    print("ç‰©ç†åŸç†è¯´æ˜ï¼š")
    print("="*60)
    print("1. æ³¢çš„å åŠ åŸç†ï¼šå½“ä¸¤ä¸ªæ³¢åœ¨åŒä¸€ä»‹è´¨ä¸­ç›¸é‡æ—¶ï¼Œå®ƒä»¬ä¼šåœ¨ç›¸é‡ç‚¹äº§ç”Ÿ")
    print("   ä½ç§»çš„çŸ¢é‡å’Œï¼Œç„¶åç»§ç»­ä¼ æ’­ï¼Œä¿æŒå„è‡ªçš„ç‰¹æ€§ä¸å˜ã€‚")
    print()
    print("2. å¹²æ¶‰ç°è±¡ï¼šä¸¤ä¸ªé¢‘ç‡ç›¸åŒã€æŒ¯åŠ¨æ–¹å‘ç›¸åŒã€ç›¸ä½å·®æ’å®šçš„æ³¢å åŠ æ—¶ï¼Œ")
    print("   æŸäº›åŒºåŸŸæŒ¯å¹…åŠ å¼ºï¼ˆç›¸é•¿å¹²æ¶‰ï¼‰ï¼ŒæŸäº›åŒºåŸŸæŒ¯å¹…å‡å¼±ï¼ˆç›¸æ¶ˆå¹²æ¶‰ï¼‰ã€‚")
    print()
    print(f"3. å½“å‰å‚æ•°åˆ†æï¼š")
    print(f"   - ç›¸ä½å·® Ï† = {phi:.2f} rad ({phi/np.pi:.2f}Ï€)")
    print(f"   - å½“ Ï† = 0, 2Ï€, 4Ï€... æ—¶ï¼Œç›¸é•¿å¹²æ¶‰æœ€æ˜æ˜¾")
    print(f"   - å½“ Ï† = Ï€, 3Ï€, 5Ï€... æ—¶ï¼Œç›¸æ¶ˆå¹²æ¶‰æœ€æ˜æ˜¾")
    print(f"   - å¹²æ¶‰å¼ºåº¦æœ€å¤§å€¼ï¼š{(A1 + A2)**2:.2f}")
    print(f"   - å¹²æ¶‰å¼ºåº¦æœ€å°å€¼ï¼š{(A1 - A2)**2:.2f}")

# åˆ›å»ºäº¤äº’ç•Œé¢
def create_interactive_app():
    # åˆ›å»ºæ»‘å—
    A1_slider = FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, 
                           description='æ³¢1æŒ¯å¹…:', continuous_update=True)
    A2_slider = FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, 
                           description='æ³¢2æŒ¯å¹…:', continuous_update=True)
    phi_slider = FloatSlider(value=0.0, min=0.0, max=2*np.pi, step=0.1,
                            description='ç›¸ä½å·®:', continuous_update=True)
    
    # æ·»åŠ æ ‡é¢˜å’Œè¯´æ˜
    title = Label(value="ğŸ“¡ æ³¢çš„å åŠ ä¸å¹²æ¶‰äº¤äº’æ¼”ç¤º")
    explanation = Label(value="è°ƒæ•´æŒ¯å¹…å’Œç›¸ä½å·®ï¼Œè§‚å¯Ÿæ³¢çš„å¹²æ¶‰ç°è±¡")
    
    # åˆ›å»ºäº¤äº’æ§ä»¶
    ui = VBox([title, explanation, A1_slider, A2_slider, phi_slider])
    
    # è¿æ¥äº¤äº’å‡½æ•°
    out = interact(wave_superposition_interference, 
                   A1=A1_slider, A2=A2_slider, phi=phi_slider)
    
    return ui, out

# è¿è¡Œåº”ç”¨
if __name__ == "__main__":
    ui, out = create_interactive_app()
    display(ui)
```
